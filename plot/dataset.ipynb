```python
from scipy import sparse

def cal_sparsity(adata):
    adata.X=sparse.csr_matrix(adata.X)
    non_zero_num=adata.X.count_nonzero()   #计算非零元素
    total_num=len(adata.obs_names)*len(adata.var_names)
    sparsity=(total_num-non_zero_num)/total_num
    return sparsity
```

```python
for i in range(50):
    
    spots=dataset[i].n_obs
    genes=dataset[i].n_vars
    sparsity=cal_sparsity(dataset[i])
    
    pd_dict['dataset'].append(i)
    pd_dict['spots'].append(spots)        
    pd_dict['genes'].append(genes)
    pd_dict['Sparsity'].append(sparsity)
    
```
