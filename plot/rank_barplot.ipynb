```python
import scanpy as sc
import numpy as np
import pandas as pd
import os
from SDMBench import *
import palettable
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd_df = pd.read_feather('performance_summary.feather')
```

```python
pd_dict_rank = {
    'DataID':[],
    'Biotech':[],
    'BiotechType':[],
    'Method':[],
    'NMI':[],
    'HOM':[],
    'COM':[],
    'CHAOS':[],
    'PAS':[],
    'ASW':[],
    'MoranI':[],
    'GearyC':[],
    'time':[],
    'memory':[]
}
```

```python
metrics_order_principle = {
    'NMI':'asc',
    'HOM':'asc',
    'COM':'asc',
    'CHAOS':'des',
    'PAS':'des',
    'ASW':'asc',
    'MoranI':'asc',
    'GearyC':'des',
    'time':'des',
    'memory':'des'
}
```

```python
data_list = np.unique(pd_df['DataID'])
```

```python
for data in data_list:
    pd_df_data = pd_df[pd_df['DataID']==data]
    num_methods = np.unique(pd_df_data['Method']).shape[0]
    # print(data,'=================')
    # print(num_methods)
    pd_dict_rank['DataID'].extend(num_methods*[data])
    pd_dict_rank['Method'].extend(np.unique(pd_df_data['Method']))
    cur_biotech = pd_df[pd_df['DataID']==data]['Biotech'].iloc[0]
    cur_biotechtype = pd_df[pd_df['DataID']==data]['BiotechType'].iloc[0]
    pd_dict_rank['Biotech'].extend(num_methods*[cur_biotech])
    pd_dict_rank['BiotechType'].extend(num_methods*[cur_biotechtype])
    
    
    for metrics in list(metrics_order_principle.keys()):
        method_score_list = []
        for method in np.unique(pd_df_data['Method']):
        
            cur_select = pd_df_data[(pd_df_data['Method']==method)]
            cur_score = np.median(cur_select[metrics])
            method_score_list.append(cur_score)
        method_score_list = np.array(method_score_list)
        # print(len(method_score_list))
        if metrics_order_principle[metrics]=='asc':
            cur_order = method_score_list.argsort().argsort()+1
        elif metrics_order_principle[metrics]=='des':
            cur_order = (-method_score_list).argsort().argsort()+1
        pd_dict_rank[metrics].extend(cur_order)
        
```

```python
pd_df_rank = pd.DataFrame(pd_dict_rank)
```

```python
pd_df_rank.to_feather('performance_summary_rank.feather')
```

```python
pd_df_rank['Accuracy'] = pd_df_rank['NMI']+pd_df_rank['HOM']+pd_df_rank['COM']
pd_df_rank['Continuity'] = pd_df_rank['CHAOS']+pd_df_rank['PAS']+pd_df_rank['ASW']
pd_df_rank['Marker'] = pd_df_rank['MoranI']+pd_df_rank['GearyC']
pd_df_rank['Scalability'] = pd_df_rank['time']+pd_df_rank['memory']
```

```python
method_order = [
    'louvain','leiden',
    'SpaGCN_without','SpaGCN_with','BayesSpace','stLearn','SEDR',
    'CCST','SCAN-IT','STAGATE','SpaceFlow','conST_nopre','BASS','DeepST'
]
```

```python
metrics_list = [
    'NMI','HOM','COM',
    'CHAOS','PAS','ASW',
    'MoranI','GearyC',
    'time','memory',
    'Accuracy','Continuity','Marker','Scalability'
]
```

```python
cmp = palettable.tableau.Tableau_20.mpl_colors
method_color_dict = dict(zip(method_order,cmp))
def make_cmp(method_color_dict,method_list):
    idx = []
    method_key = list(method_color_dict.keys())
    color_val = []
    for i in range(len(method_color_dict)):
        cur_method = method_key[i]
        cur_color = method_color_dict[cur_method]
        color_val.append(cur_color)
        if cur_method in method_list:
            idx.append(i)
    return [method_key[i] for i in idx],[color_val[i] for i in idx]
    
```

```python
for biotech in np.unique(pd_df_rank['Biotech']):
    os.makedirs(f'figures/biotech_rank/{biotech}',exist_ok=True)
    pd_df_rank_biotech = pd_df_rank[pd_df_rank['Biotech']==biotech]
    for metrics_use in metrics_list:
        cur_method_order,cur_cmp = make_cmp(method_color_dict,np.unique(pd_df_rank_biotech['Method']))
        fig,ax = plt.subplots(1,1,figsize=(2,5))

        ax = sns.barplot(pd_df_rank_biotech,x=metrics_use,y='Method',
                         # order=['spagcn','stagate','spaceflow','MS'],
                         order = cur_method_order,
                    # hue='Method',
                        ci="sd", 
                        edgecolor="black",
            errcolor="black",
            errwidth=1.5,
            capsize = 0.1,
            # alpha=0.5,

                    palette=cur_cmp
                   )

        sns.stripplot(
            pd_df_rank_biotech,
            y="Method", 
            x=metrics_use, 
                         # order=['spagcn','stagate','spaceflow','MS'],
            order = cur_method_order,

            # hue="method", 
             dodge=True, 
            # alpha=0.6, 
            ax=ax,
            color='k',
            size=2
                    # palette=cmp



        )
        handles, labels = ax.get_legend_handles_labels()
        ax.legend(handles[2:], labels[2:], title='Method', bbox_to_anchor=(1, 1.02), loc='upper left')



        plt.savefig(f'figures/biotech_rank/{biotech}/{metrics_use}.pdf',dpi=400,bbox_inches='tight',transparent=True)
        plt.close()
```
