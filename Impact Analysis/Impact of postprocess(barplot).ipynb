```python
import seaborn as sns
import pandas as pd
import os
from sklearn.metrics import *
import matplotlib.pyplot as plt
import scanpy as sc
```

```python
pd_df = pd.read_feather('/home/workspace2/zhaofangyuan/domain_output/Figure/all_data.feather')
pd_df=pd_df[pd_df.postprocess=='none'] 
pd_df
```

```python
import palettable

method_order = [
    'Louvain','Leiden',
    'SpaGCN_without','SpaGCN_with','BayesSpace','stLearn',
    'SEDR','CCST','SCAN-IT','STAGATE','SpaceFlow',
    'conST_nopre',
    'BASS',
    'DeepST'
]

dataset_order = ['151507','151508','151509','151510','151669','151670',
            '151671','151672','151673','151674','151675','151676',
            'osmfish',
            'MERFISH_0.04','MERFISH_0.09','MERFISH_0.14','MERFISH_0.19','MERFISH_0.24',
            'mouse_VISp','Slice_1','Slice_2','Slice_3',
            'global_z0','global_z1','global_z2','global_z3','global_z4',
            '20180417_BZ5_control','20180419_BZ9_control','20180424_BZ14_control','STARmap_20180505_BY3_1k']

cmp = palettable.tableau.Tableau_20.mpl_colors



method_color_dict = dict(zip(method_order,cmp))
```

```python

for n in range(len(dataset_order)):

    temp=pd_df[pd_df.dataset==dataset_order[n]] 

    p=sns.catplot(data=temp,kind="bar",x='ARI',y='algorithm',palette=method_color_dict,height=4,order=method_order)
    # p.fig.suptitle(f"{method_order[m]}_{dataset_order[n]}")
    plt.savefig(f'../postprocess_none/postprocess_ari_{dataset_order[n]}(barplot).pdf',dpi=400,bbox_inches='tight',transparent=True)
```
